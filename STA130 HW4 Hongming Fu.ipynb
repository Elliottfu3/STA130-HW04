**Question 3**
Use the percentile method. This method relies on the distribution of the bootstrap sample means. Specifically, the 95% confidence interval corresponds to the 2.5th and 97.5th percentiles of these bootstrap sample means, which captures the middle 95% of the distribution.

```python
#Question 3
import numpy as np

# Replace this with your actual data
# Example: let's assume you have a dataset of 10 numbers
data = np.array([12, 15, 14, 10, 13, 12, 14, 16, 13, 15])

# Generate bootstrap samples and calculate the mean
n_iterations = 1000  # Number of bootstrap iterations
bootstrapped_means = []  # Initialize the list to store the bootstrap means

for i in range(n_iterations):
    # Resample from the original data with replacement
    boot_sample = np.random.choice(data, size=len(data), replace=True)
    # Calculate the mean of the bootstrap sample
    bootstrapped_means.append(np.mean(boot_sample))

# Convert the bootstrap means to a numpy array
bootstrapped_means = np.array(bootstrapped_means)

# Calculate the 95% confidence interval
lower_bound = np.quantile(bootstrapped_means, 0.025)
upper_bound = np.quantile(bootstrapped_means, 0.975)

print(f"95% Bootstrap Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
#Question 4
import numpy as np

# Function to generate bootstrap samples and calculate confidence interval
def bootstrap_confidence_interval(data, num_bootstrap_samples=1000, alpha=0.05, statistic_func=np.mean):
    """
    data: Sample data (numpy array or pandas Series).
    num_bootstrap_samples: Number of bootstrap samples to generate.
    alpha: Significance level for the confidence interval (default is 0.05 for 95% CI).
    statistic_func: The function to calculate the statistic of interest (e.g., np.mean for mean, np.median for median).
    """
    # List to store the statistic for each bootstrap sample
    bootstrap_statistics = []
    
    # Generate bootstrap samples
    for _ in range(num_bootstrap_samples):
        # Resample with replacement
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        # Compute the statistic (mean by default)
        statistic = statistic_func(bootstrap_sample)
        # Store the computed statistic
        bootstrap_statistics.append(statistic)
    
    # Calculate the confidence interval
    lower_bound = np.percentile(bootstrap_statistics, 100 * (alpha / 2))
    upper_bound = np.percentile(bootstrap_statistics, 100 * (1 - alpha / 2))
    
    return lower_bound, upper_bound

# Example usage with a sample dataset (for mean)
data = np.array([10, 12, 14, 15, 18, 20, 22, 25, 30])

# Compute 95% bootstrap confidence interval for the mean
ci_mean = bootstrap_confidence_interval(data)
print("95% Bootstrap Confidence Interval for Mean:", ci_mean)

# To calculate for the median, simply change np.mean to np.median
ci_median = bootstrap_confidence_interval(data, statistic_func=np.median)
print("95% Bootstrap Confidence Interval for Median:", ci_median)

#Chat Histories https://chatgpt.com/share/66fe022e-d19c-800f-9977-fedf208321da
```
